```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.bewakoof.com/p/weekend-mood-tom--printed-half-sleeve-t-shirt'
```

```python
page=requests.get(url)
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify)
```

```python
product_name=soup.find_all('h3')
product_name 
```

```python
product_names=[]
for product_name in product_name:
    product_name=product_name.get_text().replace('\n'," ")
    product_name=product_name.strip(" ")
    product_names.append(product_name)
product_names
```

```python
print(len(product_name))
```

```python
product_name = []

for i in soup.find_all('h3'):
    product_name.append(i.text)
    
product_name   
```

```python
product_price=soup.find_all('span',class_='sellingPrice mr-1')
product_price 
```

```python
product_prices= []
for product_price in product_price:
    product_price=product_price.get_text().replace('\n'," ")
    product_price=product_price.strip(" ")
    product_prices.append(product_price)
product_prices
```

```python
print(len(product_prices))
```

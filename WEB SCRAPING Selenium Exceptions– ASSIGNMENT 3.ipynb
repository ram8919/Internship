```python
driver = webdriver.Chrome(r'c:\Users\Bawla Singh.DESKTOP-KT1PI9D\Desktop\chromedriver.exe')
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.digit.in/laptops/'
driver.get(url)
```

```python
search_button=driver.find_element_by_xpath("/html/body/div[6]/div/div[2]/div[2]/ul/li[10]/a")
search_button.click()
```

```python
title=[]
laptop_os=[]
laptop_display=[]
laptop_PROCESSOR=[]
laptop_MEMORY=[]

time.sleep(3)
#scraping the titles
titles=driver.find_elements_by_xpath("/html/body/div[5]/div[1]/div[4]/div[1]/a")
for i in titles[:10]:
    if i.text is None :
         title.append("--") 
    else:
         title.append(i.text)
        
time.sleep(3)
#scraping the laptop_os
laptops_os=driver.find_elements_by_xpath("/html/body/div[5]/div[1]/div[4]/div[2]/div[2]/div[1]/ul/li[1]/div/div")
for i in laptops_os[:10]:
    if i.text is None :
        laptop_os.append("--") 
    else:
        laptop_os.append(i.text)
        
time.sleep(3)
#scraping the laptop_display
laptop_displays=driver.find_elements_by_xpath("/html/body/div[5]/div[1]/div[4]/div[2]/div[2]/div[1]/ul/li[2]/div/div")
for i in laptop_displays[:10]:
    if i.text is None :
        laptop_display.append("--") 
    else:
        laptop_display.append(i.text)
        
time.sleep(3)
#scraping the laptop_PROCESSOR
laptop_PROCESSORS=driver.find_elements_by_xpath("/html/body/div[5]/div[1]/div[4]/div[2]/div[2]/div[1]/ul/li[4]/div/div")
for i in laptop_PROCESSORS[:10]:
    if i.text is None :
        laptop_PROCESSOR.append("--") 
    else:
        laptop_PROCESSOR.append(i.text)

time.sleep(3)
#scraping the laptop_MEMORY
laptop_MEMORYS=driver.find_elements_by_xpath("/html/body/div[5]/div[1]/div[4]/div[2]/div[2]/div[1]/ul/li[5]/div/div")
for i in laptop_MEMORYS[:10]:
    if i.text is None :
        laptop_MEMORY.append("--") 
    else:
        laptop_MEMORY.append(i.text)        
```

```python
time.sleep(3)            
 # creating the dataframe from the scraped data
df=pd.DataFrame({"title":title[:10],"laptop_os":laptop_os[:10],"laptop_display":laptop_display[:10],
                 "laptop_PROCESSOR":laptop_PROCESSOR[:10],"laptop_MEMORY":laptop_MEMORY[:10]})
df
```
